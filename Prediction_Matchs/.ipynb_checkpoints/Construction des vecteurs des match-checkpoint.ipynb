```python
#fonction qui retourne la valeur de l'attribut de la table teamAttributes en parametre pour une équipe donnée
def getTeamAttribute(team_api_id, attribute, dateMatch, train):
    # si c'est un vecteur pour le train, on peut prendre les valeurs des attributs de ce match lui meme
    if(train):
        tm = teamattributes[teamattributes["date"] <= dateMatch]
    #sinon, si c'est pour le test, on ne peut pas puisqu'on considère que le match ne s'est pas encore passé
    else:
        tm = teamattributes[teamattributes["date"] < dateMatch]
    tm = tm[tm["team_api_id"] == team_api_id]
    tm = tm[[attribute,"date"]]
    
    if(train):
        tmp = tm[tm["date"] == dateMatch]
        if(len(tmp) != 0):
            print("ici")
            tm = tmp
        
    tm = tm[[attribute]].dropna()
    
    if(len(tm) == 0):
        return 50
    else:
        return tm.mean().values[0]
    
```

```python
players_team_home = ['home_player_1', "home_player_2", "home_player_3", "home_player_4", "home_player_5",
               "home_player_6", "home_player_7", "home_player_8", "home_player_9", "home_player_10",
               "home_player_11"]

players_team_away = ["away_player_1", "away_player_2", "away_player_3", "away_player_4",
               "away_player_5", "away_player_6", "away_player_7", "away_player_8", "away_player_9",
               "away_player_10", "away_player_11"]
```

```python
def getListPlayersApiId(un_match, home):
    liste = []
    if(home):
        for pl in players_team_home:
            if pl not in un_match.columns:
                continue
            liste.append((un_match[pl].values)[0])
    else:
         for pl in players_team_away:
            if pl not in un_match.columns:
                continue
            liste.append((un_match[pl].values)[0])
    return liste
```

```python
#fonction qui retourne la valeur de l'attribut de la table playerAttributes en parametre pour une équipe donnée
def getPlayerAttribute(un_match, attribute, train, home):

    dateMatch = un_match[["date"]].values[0][0]
    list_players_api_id = getListPlayersApiId(un_match, home)
    
    moy = 0
    
    for idpl in list_players_api_id: 
        
        player=playerattributes[playerattributes.player_api_id == idpl]
        
        if(train):
            tmp = player[player["date"] == dateMatch]
            tmp = tmp[[attribute]].dropna()
            if(len(tmp) != 0):
                moy += tmp.mean().values[0]
            else:
                tmp2 = player[player["date"] <= dateMatch]
                tmp2 = tmp2[[attribute]].dropna()
                if(len(tmp2) == 0):
                    moy += 50
                else:
                    moy += tmp2.mean().values[0]
        else:
            tmp = player[player["date"] < dateMatch]
            tmp = tmp[[attribute]].dropna()
            if(len(tmp) == 0):
                moy += 50
            else:
                moy += tmp.mean().values[0]
                
    return moy*1.0 / len(list_players_api_id)
```

## Get vecteur d'une equipe

```python
def getVectMatch(un_match, train):
    vect = []
    vect2 = []
    for attribut in attribut_player:
        vect.append(getPlayerAttribute(un_match, attribut, train, True))
        vect2.append(getPlayerAttribute(un_match, attribut, train, False))
        
    dateMatch = un_match[["date"]].values[0][0]
    
    home_team_api_id = un_match[["home_team_api_id"]].values[0][0]
    away_team_api_id = un_match[["away_team_api_id"]].values[0][0]
    
    for attribut in attribut_team:
        vect.append(getTeamAttribute(home_team_api_id, attribut, dateMatch, train))
        vect2.append(getTeamAttribute(away_team_api_id, attribut, dateMatch, train))
        
    home_goal = (un_match["home_team_goal"].values)[0]
    away_goal = (un_match["away_team_goal"].values)[0]
    
    if(away_goal > home_goal):
        y = -1
    else:
        y = 1
    
    return (vect+vect2, y)
```

```python
#fonction qui construit l'ensemble de train et de test à partir de match, pour un train_split pourcent
def trainTestData(train_split):
    nbTrain = len(match) * train_split
    
    train_data = []
    train_labels = []
    
    test_data = []
    test_labels = []
    
    i = 0
    train = True
    nombre = 0
    for idm in match_id:
        print("match "+ str(nombre))
        nombre += 1
        row = match[match["match_api_id"]==idm[0]]    
        
        if(train):
            print("train")
            (vect, y) = getVectMatch(row, train)
            train_data.append(vect)
            train_labels.append(y)

            i += 1
        else:
            print("test")
            (vect, y) = getVectMatch(row, train)
            test_data.append(vect)
            test_labels.append(y)
        
        if(i >= nbTrain):
            train = False
    
    return ((train_data, train_labels),(test_data, test_labels))
```

```python
((train_data, train_labels),(test_data, test_labels)) = trainTestData(0.7)
```

```python
#on enregistre les données dans des fichiers numpy afin de pouvoir les utiliser pour l'apprentissage
np.save("./train_data", train_data)
np.save("./train_labels", train_labels)
np.save("./test_data", test_data)
np.save("./test_labels", test_labels)
```
