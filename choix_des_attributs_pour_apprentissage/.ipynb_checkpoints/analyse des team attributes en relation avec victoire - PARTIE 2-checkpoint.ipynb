# I - Description du problème

Dans ce notebook, on analyser les stratégies que suivent les différents équipes, afin de décider lesquelles doivent etre représentés dans le vecteur de notre équipe pour la phase d'apprentissage et la prédiction des résultats des match

# II - Préparation des données et Code

```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
import math
```

```python
match = pd.DataFrame(data=pd.read_csv("match.csv", sep=","))
teamattributes = pd.DataFrame(data=pd.read_csv("teamattributes.csv", sep=","))
team = pd.DataFrame(data=pd.read_csv("team.csv", sep=","))
```

```python
#Liste des matchs
match_id = np.unique(match[["match_api_id"]])
```

```python
# cette fonction retourne le nombre de victoire des match par valeur numérique de l'attribut
def attributGainEquipe(attribut):
    x = []
    for idm in match_id:
        un_match = match[match["match_api_id"]==idm]    

        home_team_id = (un_match["home_team_api_id"].values)[0]
        away_team_id = (un_match["away_team_api_id"].values)[0]    
        date = (un_match["date"].values)[0]    

        t = teamattributes[teamattributes["date"] <= date]
        
        home_goal = (un_match["home_team_goal"].values)[0]
        away_goal = (un_match["away_team_goal"].values)[0]
        
        if(home_goal > away_goal):
            tea = home_team_id
        elif(away_goal > home_goal):
            tea = away_team_id
        else:
            continue
            
        t = t[t["team_api_id"] == tea]
        
        if(len(t) == 0):
            continue
        else:
            moy = 0
            i = 0
            for index, row in t.iterrows():
                moy = moy + row[[attribut]].values[0]
                i += 1
            moy = moy*1.0 / i
            if(not np.isnan(moy)):
                x.append(moy)
    plt.hist(x)
    plt.show()
```
