# I - Description du Problème

Dans ce notebook, On va étudier quelques propriétées concernant le physique du joueur sachant que l'on a plus d'une entrée pour chaque joueur. Ceci est probablement dû aux évaluations multiples des joueurs au fil du temps, ce qui est logique, car les performances d'un joueur peuvent changer au cours de sa carrière. 


# II - Préparation des données


```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
playerattributes = pd.DataFrame(data=pd.read_csv("./donnees_soccer/playerattributes.csv", sep=","))
players=pd.DataFrame(data=pd.read_csv("./donnees_soccer/player.csv", sep=","))
players_id= np.unique(playerattributes[["player_api_id"]])
```

# III - Protocole Expérimentale

## Analyse de l'age

Pour ne pas comptre deux fois, et aussi des évaluations de la même année afin que nous puissions calculer l'âge correctement.

```python
from datetime import datetime
now = datetime.now()
# Change the players' birthday to datetime objects
players['birthday'] = pd.to_datetime(players['birthday'])

# calcul age
players['age'] = (now - players['birthday']).astype('<m8[Y]')
```

```python
sns.distplot(players['age'], hist = True, color="g")
plt.show()
```

On remarque que la plupart de joueurs l'age est entre 25 et 35 et apres le nombre diminue car il y a très peu de joueurs qui décident de continuer à jouer jusqu'à la quarantaine.

### Correlation entre Age et Potential

```python
x = []
y = []
for pl in players_id:
    player_att1=playerattributes[playerattributes.player_api_id==pl]
    player_att2=players[players.player_api_id==pl]
    x.append((player_att2["age"].values[0]))
    y.append((player_att1.sort_values(by = 'date', ascending = False)[:1]["potential"].values)[0])

```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
ax.set_xlabel("Age")
ax.set_ylabel("Potential")
plt.scatter(x, y, alpha=0.5)
plt.show()
print("Le coefficient de correlation est = ",np.corrcoef(x, y)[1,0])
```

Le coefficient de correlation est negative ainsi comme attendu le potentiel moyen de chaque groupe d'age soit le plus élevé aux âges les plus jeunes et diminue progressivement avec l'age.

### Correlation entre Age et Rating

```python
y = []
for pl in players_id:
    player_att1=playerattributes[playerattributes.player_api_id==pl]
    y.append((player_att1["overall_rating"].values)[0].mean())
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
ax.set_xlabel("Age")
ax.set_ylabel("Rating")
plt.scatter(x, y, alpha=0.5)
plt.show()
print("Le coefficient de correlation est = ",np.corrcoef(x, y)[1,0])
```

### Conclusion:

On constate un coefficient de corrélation très faible (0.07) ce qui veut dire qu'il n' y a aucune relation entre l'age du joueur et son potentiel.
